```python
# defining bonus_func for checkmating eligibility and non elegibilty
def bonus_func(sales_col):
    elig_values = np.where(sales_col >10, 'eligible', 'not eligible')
    return elig_values
# applying the bonus_func on Jan column 
store_sales['Jan_bonus']=bonus_func(store_sales['Jan'])
store_sales
```

```python
# applying the bonus_function on the store_sales to validate eligibile or not eligible
newdf = store_sales.select_dtypes(include= ['int64', 'float64']).apply(bonus_func)
newdf
```

```python
# Concatenating store_sales and Newdf dataframe column-wise
raghu = pd.concat([store_sales, newdf], axis=1)
raghu.shape
raghu.head()
```

```python
# calling the bonus_func and passing argument 6 to verify if eligibility or non-eligibiltty
bonus_func(6)
```
